```
def select_food(size, chips_dips, finger, meal, dessert, cake):
  food_dict={}
#Chips and Dips
  if chips_dips==1:
    #Chips
    chips=df3.loc[df3['Chips'] == 1]
    serving_chips=0
    while serving_chips < size:
      sample=chips.sample(n=1)
      serving_chips+=sample.serving.iloc[0]
      prod=sample.id.iloc[0]
      if prod  in food_dict.keys():
        food_dict[prod]+=1
      else:
        food_dict[prod]=1
    #Dips
    dips=df3.loc[df3['Dips'] == 1]
    serving_dips=0
    while serving_dips < size:
      sample=dips.sample(n=1)
      serving_dips+=sample.serving.iloc[0]
      prod=sample.id.iloc[0]
      if prod in food_dict.keys():
        food_dict[prod]+=1
      else:
        food_dict[prod]=1
#Finger Food
  if finger==1:
    finger_df=df3.loc[df3['Finger Food'] == 1]
    serving_finger=0
    while serving_finger < size:
      sample=finger_df.sample(n=1)
      serving_finger+=sample.serving.iloc[0]
      prod=sample.id.iloc[0]
      if prod in food_dict.keys():
        food_dict[prod]+=1
      else:
        food_dict[prod]=1
#Sit Down Meal
  if meal==1:
    meal_df=df3.loc[df3['Sit Down Meal'] == 1]
    serving_meal=0
    while serving_meal < size:
      sample=meal_df.sample(n=1)
      serving_meal+=sample.serving.iloc[0]
      prod=sample.id.iloc[0]
      if prod in food_dict.keys():
        food_dict[prod]+=1
      else:
        food_dict[prod]=1
#Dessert
  if dessert==1:
    dessert_df=df3.loc[df3['Dessert'] == 1]
    serving_dessert=0
    while serving_dessert < size:
      sample=dessert_df.sample(n=1)
      serving_dessert+=sample.serving.iloc[0]
      prod=sample.id.iloc[0]
      if prod in food_dict.keys():
        food_dict[prod]+=1
      else:
        food_dict[prod]=1
#Cake
  if cake==1:
    cake_df=df3.loc[df3['Cake'] == 1]
    serving_cake=0
    while serving_cake < size:
      sample=cake_df.sample(n=1)
      serving_cake+=sample.serving.iloc[0]
      prod=sample.id.iloc[0]
      if prod in food_dict.keys():
        food_dict[prod]+=1
      else:
        food_dict[prod]=1
        
  food_output=[]
  for i in food_dict.keys():
    price=float(df3.Price[df3['id'] == i].iloc[0][1:])
    prod_line=[ food_dict[i] , df3.Item[df3['id'] == i].iloc[0][:20] , price , (food_dict[i])*price]
    food_output.append(prod_line)
  return(food_output)    
    
    
```

```
def curate():
  guests = input("How many guests? " ) 
  party_type = input("What type of party are you hosting? " )
  food = input("What type of food do you want? [Chips & Dips, Finger Food, Sit Down Meal, Dessert, Cake] ")
  [chips_dips, finger, meal, dessert, cake] = [ food[0] , food[1] , food[2] , food[3] , food[4] ]
  drinks = input("What type of drinks do you want? [Juice, Soda, Beer, Wine, Cocktail] ")
  [juice, soda, beer, wine, cocktail] = [ drinks[0] , drinks[1] , drinks[2] , drinks[3] , drinks[4] ]
  supplies= input("What supplies do you need? [Plates, Napkins, Cups, Cutlery, Table Cloth, Serving Tools] ")
  [plates, napkins, cups, cutlery, table_cloth, serving_tools] = [supplies[0] , supplies[1] , supplies[2] , supplies[3] , supplies[4] , supplies[5]]
  decorations = input("What decorations do you want? [Photo Booth, Wall Decorations, Table Decorations, Balloons, Candles, Flowers] ")
  [photo_booth, wall, table, balloons, candles, flowers] = [decorations[0] , decorations[1] , decorations[2] , decorations[3] , decorations[4] , decorations[5]]
  run = input("Discover your kit? ")

  while run == "Yes":
    kit=pd.DataFrame(select_food(int(guests), int(chips_dips), int(finger), int(meal), int(dessert), int(cake)),columns=["Quantity", "Name", "Unit Price", "Total Price"])
    blankIndex=['']*len(kit)
    kit.index=blankIndex
    total_food=sum(list(kit["Total Price"]))
    print("\n Food: $" + str(round(total_food,2)))
    print(kit)

    kit=pd.DataFrame(select_drinks(int(guests), int(juice), int(soda), int(beer), int(wine), int(cocktail)),columns=["Quantity", "Name", "Unit Price", "Total Price"])
    blankIndex=['']*len(kit)
    kit.index=blankIndex
    total_drinks=sum(list(kit["Total Price"]))  
    print("\n Drinks: $"+str(round(total_drinks,2)))
    print(kit)

    kit=pd.DataFrame(select_supplies(int(guests), party_type, int(plates) , int(napkins) , int(cups) , int(cutlery) , int(table_cloth) , int(serving_tools)),columns=["Quantity", "Name", "Unit Price", "Total Price"])
    blankIndex=['']*len(kit)
    kit.index=blankIndex
    total_supplies=sum(list(kit["Total Price"]))  
    print("\n Supplies: $"+str(round(total_supplies,2)))
    print(kit)

    kit=pd.DataFrame(select_decorations(int(guests), party_type, int(photo_booth) , int(wall) , int(table) , int(balloons) , int(candles) , int(flowers)),columns=["Quantity", "Name", "Unit Price", "Total Price"])
    blankIndex=['']*len(kit)
    kit.index=blankIndex
    total_decorations=sum(list(kit["Total Price"]))    
    print("\n Decorations: $"+str(round(total_decorations,2)))
    print(kit) 

    print("\n Total Price: $"+str(round((total_food+total_drinks+total_supplies+total_decorations),2)) +"\n")
    run=input("Run again? ")


```
