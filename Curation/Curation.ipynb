```python
def select_food(size, chips_dips, finger, meal, dessert, cake):
    food_dict={}
#Chips and Dips
    if chips_dips==1:
    #Chips
        chips=df3.loc[df3['Chips'] == 1]
        serving_chips=0
        while serving_chips < size:
            sample=chips.sample(n=1)
            serving_chips+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
#Dips
        dips=df3.loc[df3['Dips'] == 1]
        serving_dips=0
        while serving_dips < size:
            sample=dips.sample(n=1)
            serving_dips+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
        if prod in food_dict.keys():
            food_dict[prod]+=1
        else:
            food_dict[prod]=1
#Finger Food
    if finger==1:
        finger_df=df3.loc[df3['Finger Food'] == 1]
        serving_finger=0
        while serving_finger < size:
            sample=finger_df.sample(n=1)
            serving_finger+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
#Sit Down Meal
    if meal==1:
        meal_df=df3.loc[df3['Sit Down Meal'] == 1]
        serving_meal=0
        while serving_meal < size:
            sample=meal_df.sample(n=1)
            serving_meal+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
#Dessert
    if dessert==1:
        dessert_df=df3.loc[df3['Dessert'] == 1]
        serving_dessert=0
        while serving_dessert < size:
            sample=dessert_df.sample(n=1)
            serving_dessert+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
#Cake
    if cake==1:
        cake_df=df3.loc[df3['Cake'] == 1]
        serving_cake=0
        while serving_cake < size:
            sample=cake_df.sample(n=1)
            serving_cake+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
        
    food_output=[]
    for i in food_dict.keys():
        price=float(df3.Price[df3['id'] == i].iloc[0][1:])
        prod_line=[ food_dict[i] , df3.Item[df3['id'] == i].iloc[0][:20] , price , (food_dict[i])*price]
        food_output.append(prod_line)
    return(food_output)    
    
    
```

```python
def select_supplies(size, party_type , plates , napkins , cups , cutlery , table_cloth , serving_tools):
    supplies_dict={}

#Plates  
    if plates==1 : 
        plates_df=df3.loc[df3['plates'] == 1]
        maxi = max(list(plates_df[party_type]))
        plates_maxi_df = plates_df.loc[plates_df[party_type] == maxi]               
        serving_plates=0
        while serving_plates < size:
            sample=plates_maxi_df.sample(n=1)
            serving_plates+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1

#Napkins  
    if napkins==1 : 
        napkins_df=df3.loc[df3['napkins'] == 1]
        maxi = max(list(napkins_df[party_type]))
        napkins_maxi_df = napkins_df.loc[napkins_df[party_type] == maxi]               
        serving_napkins=0
    
        while serving_napkins < size:
            sample=napkins_maxi_df.sample(n=1)
            serving_napkins+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1
               
 #Cups  
    if cups==1 : 
        cups_df=df3.loc[df3['cups'] == 1]
        maxi = max(list(cups_df[party_type]))
        cups_maxi_df = cups_df.loc[cups_df[party_type] == maxi]               
        serving_cups=0
        while serving_cups < size:
            sample=cups_maxi_df.sample(n=1)
            serving_cups+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1
#Cutlery  
    if cutlery==1 : 
        cutlery_df=df3.loc[df3['cutlery'] == 1]
        maxi = max(list(cutlery_df[party_type]))
        cutlery_maxi_df = cutlery_df.loc[cutlery_df[party_type] == maxi]               
        serving_cutlery=0
        while serving_cutlery < size:
            sample=cutlery_maxi_df.sample(n=1)
            serving_cutlery+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1
#Table Cloth  
    if table_cloth==1 : 
        tc_df=df3.loc[df3['table_cloth'] == 1]
        maxi = max(list(tc_df[party_type]))
        tc_maxi_df = tc_df.loc[tc_df[party_type] == maxi]               
        serving_tc=0

        while serving_tc < size:
            sample=tc_maxi_df.sample(n=1)
            serving_tc+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1
               
               
#Serving Tools
    if serving_tools ==1 : 
        st_df=df3.loc[df3['serving_tools'] == 1]
        maxi = max(list(st_df[party_type]))
        st_maxi_df = st_df.loc[st_df[party_type] == maxi]               
        serving_st=0

        while serving_st < size:
            sample=st_maxi_df.sample(n=1)
            serving_st+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1


    supplies_output=[]
    for i in supplies_dict.keys():
        price=float(df3.Price[df3['id'] == i].iloc[0][1:])
        prod_line=[ supplies_dict[i] , df3.Item[df3['id'] == i].iloc[0][:20] , price , (supplies_dict[i])*price]
        supplies_output.append(prod_line)
    return(supplies_output)
```

```python
from tkinter import *
class Checkbar(Frame):
    def __init__(self, parent, picks, side=LEFT, anchor=W ):
        Frame.__init__(self, parent)
        self.vars = []
        for pick in picks:
            var = IntVar()
            chk = Checkbutton(self, text=pick, variable=var)
            chk.pack(side=side, anchor=anchor, expand=YES)
            self.vars.append(var)
    def state(self):
        return map((lambda var: var.get()), self.vars)

    
    
if __name__ == '__main__':
    root = Tk()
    guests = Entry(root)
    partyType = Entry(root)
    food = Checkbar(root, ['Chips & Dips', 'Finger Food', 'Sit Down Meal', 'Dessert' , 'Cake'])
    drinks = Checkbar(root, ['Juice', 'Soda', 'Beer', 'Wine', 'Cocktail'])
    supplies = Checkbar(root, ['Plates', 'Napkins', 'Cups', 'Cutlery', 'Table Cloth', 'Serving Tools'])
    decorations = Checkbar(root, ['Photo Booth', 'Wall Decorations', 'Table Decorations', 'Balloons', 'Candles', 'Flowers'])
    
    Label(root, text = "How many guests are you hosting for? ").pack(side = TOP)
    guests.pack(side=TOP,  fill=X)
    Label(root, text = "What type of party are you hosting? ").pack()
    partyType.pack(side=TOP,  fill=X)
    Label(root, text = "What type of food do you want? ").pack()
    food.pack(side=TOP,  fill=X)
    Label(root, text = "What type of drinks do you want? ").pack()
    drinks.pack(side=TOP, fill = X)
    Label(root, text = "What type of supplies do you need? ").pack()
    supplies.pack(side=TOP, fill = X)
    Label(root, text = "What type of decorations do you want? ").pack()
    decorations.pack(side = TOP, fill = X)
    food.config(relief=GROOVE, bd=2)
    drinks.config(relief=GROOVE, bd=2)
    supplies.config(relief=GROOVE, bd=2)
    decorations.config(relief=GROOVE, bd=2)
    

    t1 = Text(root)
    
    def curate_1(): 
        t1.delete(1.0, END)
        [chips_dips, finger, meal, dessert, cake] = list(food.state())
        [juice, soda, beer, wine, cocktail] = list(drinks.state())
        [plates, napkins, cups, cutlery, table_cloth, serving_tools] = list(supplies.state())
        [photo_booth, wall, table, balloons, candles, flowers] = list(decorations.state())
        nb_guest = guests.get()
        party_type = partyType.get()
        

        class PrintToT1(object):
            def write(self,s):
                t1.insert(END,s)

        sys.stdout = PrintToT1()

        
        kit_food=pd.DataFrame(select_food(int(nb_guest), chips_dips, finger, meal, dessert, cake),columns=["Quantity", "Name", "Unit Price", "Total Price"])
        blankIndex=['']*len(kit_food)
        kit_food.index=blankIndex
        total_food=sum(list(kit_food["Total Price"]))
        print("\n Food: $" + str(round(total_food,2)))
        print(kit_food)

        kit=pd.DataFrame(select_drinks(int(nb_guest), juice, soda, beer, wine, cocktail),columns=["Quantity", "Name", "Unit Price", "Total Price"])
        blankIndex=['']*len(kit)
        kit.index=blankIndex
        total_drinks=sum(list(kit["Total Price"]))  
        print("\n Drinks: $"+str(round(total_drinks,2)))
        print(kit)

        kit=pd.DataFrame(select_supplies(int(nb_guest), party_type, plates, napkins, cups, cutlery, table_cloth, serving_tools),columns=["Quantity", "Name", "Unit Price", "Total Price"])
        blankIndex=['']*len(kit)
        kit.index=blankIndex
        total_supplies=sum(list(kit["Total Price"]))  
        print("\n Supplies: $"+str(round(total_supplies,2)))
        print(kit)

        kit=pd.DataFrame(select_decorations(int(nb_guest), party_type, photo_booth, wall, table, balloons, candles, flowers),columns=["Quantity", "Name", "Unit Price", "Total Price"])
        blankIndex=['']*len(kit)
        kit.index=blankIndex
        total_decorations=sum(list(kit["Total Price"]))    
        print("\n Decorations: $"+str(round(total_decorations,2)))
        print(kit) 

        print("\n Total Price: $"+str(round((total_food+total_drinks+total_supplies+total_decorations),2)) +"\n")

        

    Button(root, text='Discover my Kit', command=curate_1).pack(side=BOTTOM)
    t1.pack()
    t1.delete(1.0, END)

     
    root.mainloop()
```
