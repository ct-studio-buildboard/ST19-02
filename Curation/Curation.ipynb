```python
import numpy as np
import pandas as pd
df3 = pd.read_csv('Database.csv')
# Dataset is now stored in a Pandas Dataframe


def select_food(size, chips_dips, finger, meal, dessert, cake):
    food_dict={}
#Chips and Dips
    if chips_dips==1:
    #Chips
        chips=df3.loc[df3['Chips'] == 1]
        serving_chips=0
        while serving_chips < size:
            sample=chips.sample(n=1)
            serving_chips+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
#Dips
        dips=df3.loc[df3['Dips'] == 1]
        serving_dips=0
        while serving_dips < size:
            sample=dips.sample(n=1)
            serving_dips+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
        if prod in food_dict.keys():
            food_dict[prod]+=1
        else:
            food_dict[prod]=1
#Finger Food
    if finger==1:
        finger_df=df3.loc[df3['Finger Food'] == 1]
        serving_finger=0
        while serving_finger < size:
            sample=finger_df.sample(n=1)
            serving_finger+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
#Sit Down Meal
    if meal==1:
        meal_df=df3.loc[df3['Sit Down Meal'] == 1]
        serving_meal=0
        while serving_meal < size:
            sample=meal_df.sample(n=1)
            serving_meal+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
#Dessert
    if dessert==1:
        dessert_df=df3.loc[df3['Dessert'] == 1]
        serving_dessert=0
        while serving_dessert < size:
            sample=dessert_df.sample(n=1)
            serving_dessert+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
#Cake
    if cake==1:
        cake_df=df3.loc[df3['Cake'] == 1]
        serving_cake=0
        while serving_cake < size:
            sample=cake_df.sample(n=1)
            serving_cake+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in food_dict.keys():
                food_dict[prod]+=1
            else:
                food_dict[prod]=1
        
    food_output=[]
    for i in food_dict.keys():
        price=float(df3.Price[df3['id'] == i].iloc[0][1:])
        prod_line=[ food_dict[i] , df3.Item[df3['id'] == i].iloc[0][:20] , price , (food_dict[i])*price]
        food_output.append(prod_line)
    return(food_output)    
    

def select_drinks(size, juice, soda, beer, wine, cocktail):
    drinks_dict={}
#Juice
    if juice==1:
        juice_df=df3.loc[df3['Juice'] == 1]
        serving_juice=0
        while serving_juice < size:
            sample=juice_df.sample(n=1)
            serving_juice+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in drinks_dict.keys():
                drinks_dict[prod]+=1
            else:
                drinks_dict[prod]=1

#Soda
    if soda==1:
        soda_df=df3.loc[df3['Soda'] == 1]
        serving_soda=0
        while serving_soda < size:
            sample=soda_df.sample(n=1)
            serving_soda+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in drinks_dict.keys():
                drinks_dict[prod]+=1
            else:
                drinks_dict[prod]=1

#Beer
    if beer==1:
        beer_df=df3.loc[df3['Beer'] == 1]
        serving_beer=0
        while serving_beer < size:
            sample=beer_df.sample(n=1)
            serving_beer+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in drinks_dict.keys():
                drinks_dict[prod]+=1
            else:
                drinks_dict[prod]=1

#Wine
    if wine==1:
        wine_df=df3.loc[df3['Wine'] == 1]
        serving_wine=0
        while serving_wine < size:
            sample=wine_df.sample(n=1)
            serving_wine+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in drinks_dict.keys():
                drinks_dict[prod]+=1
            else:
                drinks_dict[prod]=1

#Cocktail
    if cocktail==1:
        cocktail_df=df3.loc[df3['Cocktails'] == 1]
        serving_cocktail=0
        while serving_cocktail < size:
            sample=cocktail_df.sample(n=1)
            serving_cocktail+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in drinks_dict.keys():
                drinks_dict[prod]+=1
            else:
                drinks_dict[prod]=1

        
    drinks_output=[]
    for i in drinks_dict.keys():
        price=float(df3.Price[df3['id'] == i].iloc[0][1:])
        prod_line=[ drinks_dict[i] , df3.Item[df3['id'] == i].iloc[0][:20] , price , (drinks_dict[i])*price]
        drinks_output.append(prod_line)
    return(drinks_output)    

    
def select_supplies(size, party_type , plates , napkins , cups , cutlery , table_cloth , serving_tools):
    supplies_dict={}

#Plates  
    if plates==1 : 
        plates_df=df3.loc[df3['plates'] == 1]
        maxi = max(list(plates_df[party_type]))
        plates_maxi_df = plates_df.loc[plates_df[party_type] == maxi]               
        serving_plates=0
        while serving_plates < size:
            sample=plates_maxi_df.sample(n=1)
            serving_plates+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1

#Napkins  
    if napkins==1 : 
        napkins_df=df3.loc[df3['napkins'] == 1]
        maxi = max(list(napkins_df[party_type]))
        napkins_maxi_df = napkins_df.loc[napkins_df[party_type] == maxi]               
        serving_napkins=0
    
        while serving_napkins < size:
            sample=napkins_maxi_df.sample(n=1)
            serving_napkins+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1
               
 #Cups  
    if cups==1 : 
        cups_df=df3.loc[df3['cups'] == 1]
        maxi = max(list(cups_df[party_type]))
        cups_maxi_df = cups_df.loc[cups_df[party_type] == maxi]               
        serving_cups=0
        while serving_cups < size:
            sample=cups_maxi_df.sample(n=1)
            serving_cups+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1
#Cutlery  
    if cutlery==1 : 
        cutlery_df=df3.loc[df3['cutlery'] == 1]
        maxi = max(list(cutlery_df[party_type]))
        cutlery_maxi_df = cutlery_df.loc[cutlery_df[party_type] == maxi]               
        serving_cutlery=0
        while serving_cutlery < size:
            sample=cutlery_maxi_df.sample(n=1)
            serving_cutlery+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1
#Table Cloth  
    if table_cloth==1 : 
        tc_df=df3.loc[df3['table_cloth'] == 1]
        maxi = max(list(tc_df[party_type]))
        tc_maxi_df = tc_df.loc[tc_df[party_type] == maxi]               
        serving_tc=0

        while serving_tc < size:
            sample=tc_maxi_df.sample(n=1)
            serving_tc+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1
               
               
#Serving Tools
    if serving_tools ==1 : 
        st_df=df3.loc[df3['serving_tools'] == 1]
        maxi = max(list(st_df[party_type]))
        st_maxi_df = st_df.loc[st_df[party_type] == maxi]               
        serving_st=0

        while serving_st < size:
            sample=st_maxi_df.sample(n=1)
            serving_st+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in supplies_dict.keys():
                supplies_dict[prod]+=1
            else:
                supplies_dict[prod]=1


    supplies_output=[]
    for i in supplies_dict.keys():
        price=float(df3.Price[df3['id'] == i].iloc[0][1:])
        prod_line=[ supplies_dict[i] , df3.Item[df3['id'] == i].iloc[0][:20] , price , (supplies_dict[i])*price]
        supplies_output.append(prod_line)
    return(supplies_output)


def select_decorations(size, party_type , photo_booth , wall , table , balloons , candles , flowers):
    decorations_dict={}
#Photo Booth  
    if photo_booth==1 : 
        booth_df=df3.loc[df3['photo_booth'] == 1]
        maxi = max(list(booth_df[party_type]))
        booth_maxi_df = booth_df.loc[booth_df[party_type] == maxi]               
        serving_booth=0

        while serving_booth < size:
            sample=booth_maxi_df.sample(n=1)
            serving_booth+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in decorations_dict.keys():
                decorations_dict[prod]+=1
            else:
                decorations_dict[prod]=1
#Wall Decorations  
    if wall==1 : 
        wall_df=df3.loc[df3['wall'] == 1]
        maxi = max(list(wall_df[party_type]))
        wall_maxi_df = wall_df.loc[wall_df[party_type] == maxi]               
        serving_wall=0

        while serving_wall < size:
            sample=wall_maxi_df.sample(n=1)
            serving_wall+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod  in decorations_dict.keys():
                decorations_dict[prod]+=1
            else:
                decorations_dict[prod]=1
#Table Decorations  
    if table==1 : 
        table_df=df3.loc[df3['table'] == 1]
        maxi = max(list(table_df[party_type]))
        table_maxi_df = table_df.loc[table_df[party_type] == maxi]               
        serving_table=0

        while serving_table < size:
            sample=table_maxi_df.sample(n=1)
            serving_table+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in decorations_dict.keys():
                decorations_dict[prod]+=1
            else:
                decorations_dict[prod]=1
        
#Balloons  
    if balloons==1 : 
        balloons_df=df3.loc[df3['balloons'] == 1]
        maxi = max(list(balloons_df[party_type]))
        balloons_maxi_df = balloons_df.loc[balloons_df[party_type] == maxi]               
        serving_balloons=0

        while serving_balloons < size:
            sample=balloons_maxi_df.sample(n=1)
            serving_balloons+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in decorations_dict.keys():
                decorations_dict[prod]+=1
            else:
                decorations_dict[prod]=1
        
#Candles  
    if candles==1 : 
        candles_df=df3.loc[df3['candles'] == 1]
        maxi = max(list(candles_df[party_type]))
        candles_maxi_df = candles_df.loc[candles_df[party_type] == maxi]               
        serving_candles=0

        while serving_candles < size:
            sample=candles_maxi_df.sample(n=1)
            serving_candles+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in decorations_dict.keys():
                decorations_dict[prod]+=1
            else:
                decorations_dict[prod]=1
        
#Flowers  
    if flowers==1 : 
        flowers_df=df3.loc[df3['flowers'] == 1]
        maxi = max(list(flowers_df[party_type]))
        flowers_maxi_df = flowers_df.loc[flowers_df[party_type] == maxi]               
        serving_flowers=0

        while serving_flowers < size:
            sample=flowers_maxi_df.sample(n=1)
            serving_flowers+=sample.serving.iloc[0]
            prod=sample.id.iloc[0]
            if prod in decorations_dict.keys():
                decorations_dict[prod]+=1
            else:
                decorations_dict[prod]=1


    decorations_output=[]
    for i in decorations_dict.keys():
        price=float(df3.Price[df3['id'] == i].iloc[0][1:])
        prod_line=[ decorations_dict[i] , df3.Item[df3['id'] == i].iloc[0][:20] , price , (decorations_dict[i])*price]
        decorations_output.append(prod_line)
    return(decorations_output)
```
